In the following example we try to use more visualization features to save the heatmaps. However, you can make the heatmap smaller by increasing the `var_cutoff` threshold.

```python
import pylab
import matplotlib.pyplot as plt
import seaborn as sns
annot_kws={"size": 1}
fig = gps.plot_heatmap(var_cutoff=0.2, figsize = (50, 4.5))
ax_heatmap = fig.ax_heatmap
ax_heatmap.set_xlabel('Omics features', fontweight='bold', fontsize=10)
ax_heatmap.set_ylabel('Dyanmaics ', fontweight='bold', fontsize=10)
ax_heatmap.legend(framealpha = 0.2,prop={'size':1}) 
ax_heatmap.get_xaxis().set_tick_params(which='both', labelsize=7)# top='off', bottom='on', direction='out')
ax_heatmap.get_yaxis().set_tick_params(which='both', labelsize=6)
# You may also want save the plots in '.pdf' or vector format like '.svg'
plt.savefig('/Users/alireza/DataspellProjects//waveome/examples/Sleeve_surgery/000Lipid_heatmap.svg', dpi=600, bbox_inches = "tight") 
plt.savefig('/Users/alireza/DataspellProjects//waveome/examples/Sleeve_surgery/000Lipid_heatmap.pdf', dpi=600, bbox_inches = "tight") 
plt.show()
```
